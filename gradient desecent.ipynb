```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ins.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['age', 'affordibility']], df.bought_insurance, test_size=0.2)
```

```python
len(x_train)
```

```python
x_train_scaled = x_train.copy()
x_train_scaled['age'] = x_train_scaled['age']/100
x_test_scaled = x_test.copy()
x_test_scaled['age'] = x_test_scaled['age']/100
```

```python
x_train_scaled
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train_scaled, y_train, epochs=5000)
```

```python
model.evaluate(x_test_scaled, y_test)
```

```python
def logloss(y_true, y_predicted):
    ep = 1e-15
    y_predicted_new = [max(i, ep) for i in y_predicted]
    y_predicted_new = [min(i, 1-ep) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sig(x):
    import math
    return 1/(1+np.exp(-x))
sig(np.array([12, 0, 1]))
```

```python
def gradient_dec(age, aff, y_true, epochs, loss_thres=0.4631):
    #w1, W2, bias
    w1=w2=1
    bias=0
    rate=0.1
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * aff + bias#all vector multiplaction in one shot .. coz of numpy
        y_predicted = sig(weighted_sum)
        loss = logloss(y_true, y_predicted)
        w1d = (1/n)*np.dot(np.transpose(age), (y_predicted - y_true))
        w2d = (1/n)*np.dot(np.transpose(aff), (y_predicted - y_true))
        biasd = np.mean(y_predicted-y_true)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * biasd
        print(f'epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        if loss <= 0.4631:
            break
    return w1, w2, bias
```
